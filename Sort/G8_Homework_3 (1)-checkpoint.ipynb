```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

sns.set()
rand_state = 1000
```

Read in data

```python
df = pd.read_csv('UCI_Credit_Card.csv')
df_raw = df
df.head()
```

#### Cleaning

```python
df.info()
```

The variable 'PAY_0' is the payment status in September. The 'BILL_AMT1' and 'PAY_AMT1' variables are both for September, so 'PAY_0' should be changed to 'PAY_1' to avoid confusion.

```python
df.rename(columns = {'PAY_0':'PAY_1'}, inplace = True)
```

This next chunk is renaming variable types.

```python
categoricals = ['SEX', 'AGE', 'EDUCATION', 'MARRIAGE', 'PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5',
                'PAY_6']
numerics = ['ID', 'LIMIT_BAL', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4',
            'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

for col in categoricals:
    df[col] = df[col].astype("category")
    
for col in numerics:
    df[col] = pd.to_numeric(df[col], errors = 'coerce')
```

```python
df.info()
```

#### Null Values

```python
df.replace(' ', np.nan, inplace = True)
df.isnull().sum()/ len(df) * 100
```

```python
sum(df.duplicated())
```

There are no null values, or duplicate values.

#### More Cleaning

```python
df.describe().T
```

Everything seems to be fine in the describe.

Default payment variables needs to be changed to make visualization easier. We also replaced any values of 6 in education to 5, because both represent "unknown" schooling. There are also 14 observations where education is '0', we didn't want to just drop these values, so we made them 'unknown' as well.

```python
df.rename(columns={'default.payment.next.month':'def_pay'}, inplace=True)
df['EDUCATION'] = df['EDUCATION'].replace(6,5) 
df['EDUCATION'] = df['EDUCATION'].replace(0,5) 
df.head()
```

The last thing we'll do is drop the ID variable as it is not useful for our analysis.

```python
df.drop(['ID'], axis = 1, inplace = True)
df.head()
```

#### Visualization

First we wanted to find a way to visualize the target variable. We decided to make a barplot for the default variable.

```python
def_cnt = (df.def_pay.value_counts(normalize=True)*100)
def_cnt.plot.bar(figsize=(6,6))
plt.xticks(fontsize=12, rotation=0)
plt.yticks(fontsize=12)
plt.title("Probability Of Defaulting Payment Next Month", fontsize=15)
for x,y in zip([0,1],def_cnt):
    plt.text(x,y,y,fontsize=12)
plt.show()
```

0 is for non defaulted accounts, 1 is for defaulted accounts. As can be seen, about 78% of the observations in this dataset are accounts that did not default the following month. 22% of the accounts in this dataset defaulted the following month. 

----

Next we created a bar chart that showed the proportion of accounts that defaulted based on the level of education.

```python
target_count = df.groupby(['EDUCATION','def_pay']).EDUCATION.count().unstack()

# print(target_count)
plt.figure(figsize=(6,6))
target_count.plot(kind='bar', stacked=True)
plt.xticks([0,1,2,3,4], labels=["Grad School", "University", "High School", "Other","Unknown"], rotation=45)
plt.ylabel('Count')
plt.legend(['Not defaulted', 'Defaulted'])
plt.title("Education level and People deafulted counts")
plt.show()
```

As can be seen, university has the most observations, with grad school second, and high school third. Other and unknown both have very few observations. The proportion for each education level follows a similar proportion as the dataset as a whole, most accounts do not default, with only a relatively small proportion defaulting. The proportion defaulted for high school education is actually quite small which is interesting. We thought it was interesting that university education had the most observations. We thought this might be because most people who have a university education are probably in situations financially where they are applying and getting accepted for more credit loans. The other education is people who dont even have a high school education, and usually these kind of people are in pretty desperate financial situations and probably not getting approved for many credit card loans.

----

The last visualization is a distribution plot for LIMIT_BAL, which is the amount of given credit in New Taiwan dollars.

```python
plt.subplots(figsize=(10,5))
# plt.subplot(121)
sns.distplot(df.LIMIT_BAL)
plt.show()
```

As can be seen, this is a right skewed distribution. This makes sense, most credit limits are low in amount, with only a few individuals getting high credit limits. We see that the highest peak is around 100,000 NT dollars, with smaller peaks following as the credit limit is increased.

```python
plt.figure(figsize=(12,6))

sns.boxplot(x = "SEX", y = "LIMIT_BAL",data = df, palette = 'rocket', showmeans=True, 
            meanprops={"markerfacecolor":"red",  "markeredgecolor":"black", "markersize":"10"})

plt.ticklabel_format(style='plain', axis='y') #repressing scientific notation    
plt.xticks([0,1],['Male', 'Female'], fontsize = 12)

plt.show()
```

We decided to use a heat map to check correlations between the variables. This is only for the numeric variables.

```python
plt.subplots(figsize=(30,20))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

It looks like there is NOT a strong correlation with def_pay and any of the other numeric variables. Most of the variable correlations are positive. The strongest correlations are between bill amounts. The strongest of these are between consecutive months. This makes sense, if there is a really high bill one month, its very likely that a person will pay the minimum amount and that bill will stay high the next month. The strength of the correlation falls the further from the base month you get.

----

Lastly, we decided to do a pairwise plot to get a better look at the relationships among the features.

```python
sns.pairplot(df)
```

 Similar to the heatmap, bill amounts are strongly correlated. In these plots there appear to be some variation in the smaller dollar amounts then the variance is a bit more constant the higher the dollar amount is. Looking at the credit limit vs bill and pay amount plots, we see that usually lower credit limits have small variation in the amount billed or paid in a month, the higher credit limits vary greatly on the amount they are billed or the amount they pay. None of these plots appear to be that linear.
